```python
import pandas as pd

CSV_PATH = "./Experimental_Data/llava_pastamm_mme_gain_vs_baseline_0-21.csv"

# ===== 读取数据 =====
df = pd.read_csv(CSV_PATH)

# ===== 找出所有 gain 指标列 =====
gain_cols = [col for col in df.columns if col.endswith("_gain")]

# ===== 在所有 layer-head 上求和 =====
gain_sum = df[gain_cols].sum()

# ===== 转成 DataFrame 方便排序和显示 =====
gain_sum_df = gain_sum.reset_index()
gain_sum_df.columns = ["metric", "total_gain_sum"]

# ===== 按总涨幅排序 =====
gain_sum_df = gain_sum_df.sort_values(
    by="total_gain_sum",
    ascending=False
).reset_index(drop=True)

# ===== 打印结果 =====
print("=== Gain Sum Ranking (All Heads & Layers) ===")
print(gain_sum_df)

# ===== 单独查看最优 & 最差 =====
print("\n=== Top Positive Gain Metric ===")
print(gain_sum_df.iloc[0])

print("\n=== Most Negative Gain Metric ===")
print(gain_sum_df.iloc[-1])
```
